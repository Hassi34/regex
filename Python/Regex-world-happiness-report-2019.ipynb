```python
import pandas as pd 
import re 
```

```python
df = pd.read_csv('./world-happiness-report-2019.csv')
df
```

```python
#Extract first five letters form a str
df['first_five_Letter']=df['Country (region)'].str.extract(r'(^.{0,5})')
df
```

```python
df[df['Country (region)'].str.count(r'(^f.)', re.IGNORECASE) ==1]
```

```python
df['Country (region)'].str.count(r'(^F.)').to_frame().value_counts()
```

```python
df['Country (region)'].str.count(r'(^f)', re.IGNORECASE).sum()
```

```python
df[df['Country (region)'].str.count(r'(^p.)', re.IGNORECASE)==1]
```

```python
df[df['Country (region)'].str.match(r"^P.")==True]
```

```python
# Contains
df[df['Country (region)'].str.contains(r"(^I)") == True]
```

```python
#combining two columns to create a str
new_str = df[['Country (region)','first_five_Letter']].agg(' '.join, axis=1)
new_str
```

```python
#pandas split
new_str.str.split("i", expand=True)
```

```python
#spliting on space 
new_str.str.split("\s", expand=True, n=-1)
```
